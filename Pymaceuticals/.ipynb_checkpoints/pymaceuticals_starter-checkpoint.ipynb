### Analysis

- Observations or inferences that can be made from the data:

1. Based on the box plot of the final tumor volume for the top four drug regimens (Capomulin, Ramicane, Infubinol, and Ceftamin), it can be observed that Capomulin and Ramicane are more effective at reducing tumor volume compared to Infubinol and Ceftamin. Both Capomulin and Ramicane have a smaller median tumor volume and smaller range of values compared to the other two drug regimens. However, it is to be noted that Ramicane has lower median of tumor volume than Capomulin.

2. Based on the scatter plot of tumor volume versus mouse weight for the Capomulin treatment regimen, the correlation between mouse weight and the average tumor volume appears to be strong, with a correlation coefficient of 0.84. This indicates a positive linear relationship between these two variables, suggesting that as the weight of the mouse increases, so does the average tumor volume. However, the scatter plot also shows that the majority of the mice treated with Capomulin had an average tumor volume of less than 50 cubic millimeters.

3. The bar charts of the total number of time points for all mice tested for each drug regimen throughout the study show that Capomulin and Ramicane had the most data points, while Propriva had the fewest. This may suggest that Capomulin and Ramicane were more heavily studied than other drug regimens and could be more promising candidates for further research. It is also interesting to note that the number of time points for the other drugs falls somewhere in between, with Ketapril, Naftisol, Zoniferol, Placebo, Stelasyn, Ceftamin, and Infubinol all having similar numbers of time points.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_df=pd.merge(study_results,mouse_metadata,on='Mouse ID',how='outer')
# Display the data table for preview
merged_df
```

```python
# Checking the number of mice.
num_unique_mice=merged_df['Mouse ID'].nunique()
num_unique_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse=(merged_df.loc[merged_df.duplicated(subset=['Mouse ID','Timepoint']),'Mouse ID']).unique()
duplicate_mouse
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_data=merged_df.loc[merged_df['Mouse ID']==duplicate_mouse[0],:]
duplicate_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_df = merged_df.loc[merged_df['Mouse ID']!=duplicate_mouse[0]]
cleaned_df
```

```python
# Checking the number of mice in the clean DataFrame.
num_unique_mice = cleaned_df["Mouse ID"].nunique()
print(f"Number of unique mice IDs after removing duplicate Mouse ID: {num_unique_mice}")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

grouped_df=cleaned_df.groupby('Drug Regimen')
mean_tumor_volume=grouped_df['Tumor Volume (mm3)'].mean()
median_tumor_volume=grouped_df['Tumor Volume (mm3)'].median()
var_tumor_volume=grouped_df['Tumor Volume (mm3)'].var()
std_tumor_volume=grouped_df['Tumor Volume (mm3)'].std()
sem_tumor_volume=grouped_df['Tumor Volume (mm3)'].sem()
summary_df=pd.DataFrame({'Mean Tumor Volume':mean_tumor_volume,
                        'Median Tumor Volume':median_tumor_volume,
                        'Tumor Volume Variance':var_tumor_volume,
                        'Tumor Volume Std. Dev.':std_tumor_volume,
                        'Tumor Volume Std. Err.':sem_tumor_volume})
summary_df


```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
summary_df = grouped_df.agg({"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]})
summary_df.columns = [" ".join(col) for col in summary_df.columns]
renamed_summary_df=summary_df.rename(columns={'Tumor Volume (mm3) mean':'Mean Tumor Volume',
                                             "Tumor Volume (mm3) median":"Median Tumor Volume",
                                             'Tumor Volume (mm3) var':'Tumor Volume Variance',
                                             'Tumor Volume (mm3) std':'Tumor Volume Std. Dev.',
                                             'Tumor Volume (mm3) sem': 'Tumor Volume Std. Err.'})
renamed_summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse

last_timepoint = cleaned_df.groupby("Mouse ID")["Timepoint"].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_last_timepoint = pd.merge(last_timepoint, cleaned_df, on=["Mouse ID", "Timepoint"], how="left")
merged_last_timepoint
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for regimen in treatment_regimens:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volume = merged_last_timepoint.loc[merged_last_timepoint["Drug Regimen"] == regimen, "Tumor Volume (mm3)"]
    # add subset 
    tumor_volume_data.append(tumor_volume)
    # Determine outliers using upper and lower bounds
    quartiles = tumor_volume.quantile([0.25,0.5,0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75] 

    iqr = upperq - lowerq
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    outliers = tumor_volume.loc[(tumor_volume < lower_bound) | (tumor_volume > upper_bound)]
    print(f"{regimen}'s potential outliers: {outliers}")
    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig,ax=plt.subplots()
ax.boxplot(tumor_volume_data,labels=treatment_regimens, flierprops={"markerfacecolor": "red", "markersize": 10})
plt.ylabel("Final Tumor Volume (mm3)")
plt.title("Final Tumor Volume for Most Promising Treatment Regimens")
plt.legend(["Outliers"])
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
# Select a mouse that was treated with Capomulin
capomulin_data = cleaned_df.loc[cleaned_df["Drug Regimen"] == "Capomulin"]
capomulin_data
# Select a mouse that was treated with Capomulin
mouse_id = "l509"
mouse_data = capomulin_data.loc[capomulin_data["Mouse ID"] == mouse_id]

# Generate line plot of tumor volume versus time point for the selected mouse
plt.plot(mouse_data["Timepoint"], mouse_data["Tumor Volume (mm3)"])
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title(f"Tumor Volume Over Time for Mouse {mouse_id}")
plt.legend(["Mouse ID: " + mouse_id])
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
# Select data for Capomulin treatment regimen
capomulin_data = cleaned_df.loc[cleaned_df["Drug Regimen"] == "Capomulin"]

# Calculate average tumor volume and mouse weight for each mouse
avg_tumor_vol = capomulin_data.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
mouse_weight = capomulin_data.groupby("Mouse ID")["Weight (g)"].mean()

# Create a new DataFrame with the average tumor volume and mouse weight for each mouse
scatter_data = pd.DataFrame({"Average Tumor Volume (mm3)": avg_tumor_vol, "Weight (g)": mouse_weight})

# Generate scatter plot of tumor volume versus mouse weight for the Capomulin treatment regimen
scatter_data.plot(kind="scatter", x="Weight (g)", y="Average Tumor Volume (mm3)")
plt.title("Tumor Volume vs. Mouse Weight for Capomulin Treatment")
plt.legend(["Mouse Data"])
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
from scipy.stats import linregress
(slope, intercept, rvalue, pvalue, stderr) = linregress(scatter_data["Weight (g)"], scatter_data["Average Tumor Volume (mm3)"])
print(f'The correlation between mouse weight and the average tumor volume is {round(rvalue,2)}')
regress_values = scatter_data["Weight (g)"] * slope + intercept
scatter_data.plot(kind="scatter", x="Weight (g)", y="Average Tumor Volume (mm3)")
plt.plot(scatter_data["Weight (g)"],regress_values,color='red')
eqn = f'y = {slope:.2f}x + {intercept:.2f}'
plt.annotate(eqn,xy=(20,36))
plt.title("Tumor Volume vs. Mouse Weight for Capomulin Treatment")
plt.legend(["Mouse Data"])
plt.show()
```
